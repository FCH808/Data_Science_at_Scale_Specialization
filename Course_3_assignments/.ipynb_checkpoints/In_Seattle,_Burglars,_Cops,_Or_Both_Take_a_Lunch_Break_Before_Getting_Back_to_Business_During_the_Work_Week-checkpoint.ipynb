# In Seattle, Burglars, Cops, Or Both Take a Lunch Break Before Getting Back to Business During the Work Week

## Load Data and Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
sf_data = pd.read_csv('sanfrancisco_incidents_summer_2014.csv',
                      parse_dates=['Date', 'Time'],
                      infer_datetime_format=True)

seattle_data = pd.read_csv('seattle_incidents_summer_2014.csv', 
                           parse_dates=['Date Reported', 
                                        'Occurred Date or Date Range Start',
                                        'Occurred Date Range End'],
                           infer_datetime_format=True)
```

## Data formatting and cleanup

___First off, right away we need to remove 1189 records that were included in the Seattle Data which were outside the summer of 2014 (06/01/2014  - 08/31/2014)___

```python
seattle_data = seattle_data[seattle_data['Date Reported'] <= '2014-08-31']
```

```python
# Make Hour and FullDayOfWeek categorical variables to have groupby
#  leave empty groups/levels in the dataset; others were not used in this visualization
seattle_data['Time'] = seattle_data['Date Reported'].dt.time
seattle_data['Day'] = seattle_data['Date Reported'].dt.day.astype('category')
seattle_data['Year'] = seattle_data['Date Reported'].dt.year.astype('category')
seattle_data['Month'] = seattle_data['Date Reported'].dt.month.astype('category')
seattle_data['DayOfWeek'] = seattle_data['Date Reported'].dt.dayofweek.astype('category')
seattle_data['Date'] = seattle_data['Date Reported'].dt.date
seattle_data['Hour'] = seattle_data['Date Reported'].dt.hour.astype('category')
seattle_data['Week'] = seattle_data['Date Reported'].dt.weekofyear.astype('category')

day_names = {0: 'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday',
             4:'Friday', 5:'Saturday', 6:'Sunday'}

seattle_data['FullDayOfWeek'] = seattle_data['DayOfWeek'].apply(lambda x: day_names[x])
seattle_data['FullDayOfWeek'] = seattle_data['FullDayOfWeek'].astype('category')

# Rename overall description for consistent names across datasets
seattle_data = seattle_data.rename(columns={'Summarized Offense Description': 'Category'})
```

```python
sf_data['Day'] = sf_data['Date'].dt.day.astype('category')
sf_data['Year'] = sf_data['Date'].dt.year.astype('category')
sf_data['Month'] = sf_data['Date'].dt.month.astype('category')
sf_data['DayOfWeek'] = sf_data['Date'].dt.dayofweek.astype('category')
#sf_data['Date'] = sf_data['Date'].dt.date
sf_data['Hour'] = sf_data['Time'].dt.hour.astype('category')
sf_data['Week'] = sf_data['Date'].dt.weekofyear.astype('category')

sf_data['FullDayOfWeek'] = sf_data['DayOfWeek'].apply(lambda x: day_names[x])
sf_data['FullDayOfWeek'] = sf_data['FullDayOfWeek'].astype('category')
```

## Define plotting functions

```python
def plot_hourly_crime(data, crime='BURGLARY'):
    
    # Subset out the passed in crime.
    subset_data = data[data['Category'] == crime]

    # Group by Hour and Day of the Week.
    hourly_subset = subset_data.groupby(['FullDayOfWeek', 'Hour'], as_index=True).count().reset_index()
    
    # Oddly, one data set coerced datatypes, while the other retained 'category' type status.
    # Explicitly,  change data types before filling in any missing values
    hourly_subset['Hour'] = hourly_subset['Hour'].astype('int64')
    hourly_subset['Category'] = hourly_subset['Category'].astype('float64')
    hourly_subset['FullDayOfWeek'] = hourly_subset['FullDayOfWeek'].astype('object')
    
    # Fill in NaN's with 0 if there are any that came up counting over sub-groups
    hourly_subset = hourly_subset[['Hour', 'Category', 'FullDayOfWeek']].fillna(0)
    
    hourly_subset = hourly_subset.rename(columns={'Category': 'Count'})

    # Get the averages of the hourly data.
    avg_hourly = hourly_subset.groupby(['Hour'], as_index=False).mean()
    
    avg_hourly['FullDayOfWeek'] = 'Average'

    # Add in Averages as their own entry
    full_hourly = pd.concat([hourly_subset, avg_hourly], axis=0)

    # Specify the order of days to be plotted
    day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday',
                 'Friday', 'Saturday', 'Sunday']
    
    col_order = ['Average'] + day_names
    # Reformat the time for the plot
    time_labels = ['%02d:00' % x for x in np.arange(24)]
    
    # Plot the crime
    grid = sns.factorplot(data=full_hourly, 
                          x='Hour', y='Count',
                          hue='FullDayOfWeek',
                          col='FullDayOfWeek',
                          col_wrap=4,
                          col_order=col_order,
                          size=5, palette='Set2')
    # Add labels/rotate x-axis
    grid.set_xticklabels(labels=time_labels, rotation=90)
    grid.set(ylim=(0, None)) # Start y-axis at 0, no neg. crime counts
    plt.subplots_adjust(top=0.93) # Make room for main title
    grid.fig.suptitle('{} Counts for Each Day of the Week Per Hour (Avg. included)'.format(crime.title()), size=15)
```

```python
def plot_top10():
    
    top10_seattle = pd.DataFrame({'Counts':seattle_data.Category.value_counts()[:10]})
    top10_seattle['Category'] = top10_seattle.index
    
    top10_sf = pd.DataFrame({'Counts':sf_data.Category.value_counts()[:10]})
    top10_sf['Category'] = top10_sf.index

    fig = plt.figure()
    plt.figure(figsize=(12, 8))
    plt.subplot(2,1,1)
    plt.bar(np.arange(10), top10_seattle.Counts)
    plt.xticks(np.arange(10), top10_seattle.index, rotation=45)
    plt.ylim(0, 10000)
    plt.ylabel('Counts')
    plt.title('Top 10 Crime Counts for Seattle - Summer of 2014')

    plt.subplot(2,1,2)
    plt.bar(np.arange(10), top10_sf.Counts)
    plt.xticks(np.arange(10), top10_sf.index, rotation=45)
    plt.ylabel('Counts')
    plt.title('Top 10 Crime Counts for San Francisco - Summer of 2014')

    plt.tight_layout()
    
```

## Analysis

### Overall Numbers

First, let's take a quick look at the top 10 reported crimes in Seattle and San Francisco during the summer of 2014 broken down by hourly counts for each day of the week.

In total, there were 28,993 reported crimes in the San Francisco dataset, and 31,590 in the Seattle dataset.

```python
plot_top10()
```

Right away, we can see that crime is reported differently with different overlapping categories. 

A direct comparison will be difficult without extra data munging outside the scope of this analysis.

But we can get an overall feel for the distributions here.

---

Car prowling is the most frequent past-time (category) for criminals in Seattle, while San Franciscan criminals by far enjoy partaking in the combined category of 'Larceny/Theft.'

## A Closer Look - Drilling Down

In seattle, there seems to be a interesting lunchtime lull on several days of the week (not including weekends) for burglars in particular.

```python
plot_hourly_crime(seattle_data, 'BURGLARY')
```

Here, we look the burglaries per hour for each day of the week in Seattle.

On Monday, Wednesday, and Thursday there is a sharp decline from 11am to 12pm. Even on Friday there is a decline around this time, while Tuesdays and Saturdays seems to be late lunch day (perhaps Taco Tuesday 2pm specials galore!)

This is such a large decline, that is markedly affects the hourly averages over all days!

It's not clear though if this criminals getting carbs for the second half the day, or police grabbing some calories for future arresting (or maybe both in a lunch-time truce!)

---

If we look at the most similar category for San Francisco, 'Larceny/Theft, we don't quite see phenomenon.

```python
plot_hourly_crime(sf_data, 'LARCENY/THEFT')
```

A key take-away here is consistency - roll out of bed around 5am, and slowly starting ramping up thieving as the day progresses until the 5-6pm rush before ramping down around midnight during the week.

Saturday is the day of opportunity though, peaking at midnight, before returning to business-as-usual on Sunday.

## Conclusion 

There are some interesting trends in hourly/daily crimes in the data. When it comes to burglary, Seattle seems to take their food seriously, while San Francisco keeps a strong consistent work ethic, putting in extra long hours on the weekend.
